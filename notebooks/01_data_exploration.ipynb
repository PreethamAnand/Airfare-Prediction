```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load data
df = pd.read_csv('../data/raw/flight_prices.csv')

# Convert duration to minutes
df['duration'] = df['duration'].apply(lambda x: int(float(x)*60))
```

```python

# 1. Price Distribution
plt.figure(figsize=(10,6))
sns.histplot(df['price'], kde=True, bins=30)
plt.title('Flight Price Distribution')
plt.xlabel('Price (INR)')
plt.ylabel('Count')
plt.savefig('../reports/figures/price_distribution.png')
plt.show()

```

```python
# 2. Price by Airline
plt.figure(figsize=(12,6))
sns.boxplot(x='airline', y='price', data=df)
plt.xticks(rotation=45)
plt.title('Price Distribution by Airline')
plt.savefig('../reports/figures/price_by_airline.png')
plt.show()
```

```python

# 3. Price vs Days Left
plt.figure(figsize=(10,6))
sns.scatterplot(x='days_left', y='price', data=df, alpha=0.5)
plt.title('Price vs Days Until Departure')
plt.xlabel('Days Until Departure')
plt.ylabel('Price (INR)')
plt.savefig('../reports/figures/price_vs_days_left.png')
plt.show()
```

```python

# 4. Price by Number of Stops
plt.figure(figsize=(8,6))
sns.boxplot(x='stops', y='price', data=df, order=['zero', 'one', 'two_or_more'])
plt.title('Price Distribution by Number of Stops')
plt.savefig('../reports/figures/price_by_stops.png')
plt.show()
```
